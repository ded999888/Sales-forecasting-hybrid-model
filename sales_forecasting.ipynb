```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestRegressor
from sklearn.preprocessing import OneHotEncoder, MinMaxScaler
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
from tensorflow.keras.optimizers import Adam

# Шаг 1: Подготовка данных
# ------------------------

# Загрузка данных
data = pd.read_excel("/content/drive/MyDrive/Питьевая вода для прогноза.xlsx")  # Убедитесь, что файл существует

# Кодирование категориальных переменных ("Промо", "Праздник", "Сезонность")
encoder = OneHotEncoder()
categorical_features = encoder.fit_transform(data[["Промо", "Праздник", "Сезонность"]]).toarray()

# Извлечение числовых признаков
numeric_features = data[["Цена", "Единицы", "Объем продаж"]].values

# Объединение категориальных и числовых данных
processed_data = np.hstack((numeric_features, categorical_features))

# Нормализация данных
scaler = MinMaxScaler()
processed_data = scaler.fit_transform(processed_data)

# Преобразование данных в последовательности
def create_sequences(data, time_steps=7):
    X, y = [], []
    for i in range(len(data) - time_steps):
        X.append(data[i:i+time_steps])  # Последовательности для моделей
        y.append(data[i+time_steps, 2])  # Целевая переменная (Объём продаж)
    return np.array(X), np.array(y)

time_steps = 7
X_lstm, y_lstm = create_sequences(processed_data, time_steps)

# Разделение данных на обучающую и тестовую выборки
train_size = int(0.8 * len(X_lstm))
X_train, X_test = X_lstm[:train_size], X_lstm[train_size:]
y_train, y_test = y_lstm[:train_size], y_lstm[train_size:]

# Шаг 2: Обучение модели LSTM
# ---------------------------

# Создание модели LSTM
lstm_model = Sequential([
    LSTM(64, input_shape=(time_steps, X_lstm.shape[2]), return_sequences=True),
    Dropout(0.2),
    LSTM(96),
    Dropout(0.2),
    Dense(1)  # Один выход
])
lstm_model.compile(optimizer=Adam(learning_rate=0.00087434), loss='mean_squared_error')

# Обучение модели
history = lstm_model.fit(X_train, y_train, epochs=20, batch_size=32, validation_split=0.2)

# Прогноз с использованием LSTM
lstm_forecast = lstm_model.predict(X_test)
lstm_forecast_flat = lstm_forecast.flatten()

# Шаг 3: Обучение модели Random Forest
# ------------------------------------

# Преобразование данных для Random Forest
rf_input = np.hstack((
    lstm_forecast_flat.reshape(-1, 1),
    encoder.transform(data[["Промо", "Праздник", "Сезонность"]].iloc[-len(y_test):]).toarray()
))

# Обучение Random Forest
rf_model = RandomForestRegressor(n_estimators=100, random_state=42)
rf_model.fit(rf_input, y_test)

# Прогноз с использованием Random Forest
rf_forecast = rf_model.predict(rf_input)

# Шаг 4: Оценка качества моделей
# ------------------------------

def evaluate_model(true_values, predictions, model_name="Модель"):
    """
    Функция для оценки модели с использованием метрик MAE, RMSE и R^2.
    """
    mae = mean_absolute_error(true_values, predictions)
    mse = mean_squared_error(true_values, predictions)
    rmse = np.sqrt(mse)
    r2 = r2_score(true_values, predictions)

    print(f"Оценка модели {model_name}:")
    print(f"  Средняя абсолютная ошибка (MAE): {mae:.2f}")
    print(f"  Корень из среднеквадратичной ошибки (RMSE): {rmse:.2f}")
    print(f"  Коэффициент детерминации (R^2): {r2:.2f}")
    return mae, rmse, r2

# Оценка моделей
evaluate_model(y_test, lstm_forecast_flat, model_name="LSTM")
evaluate_model(y_test, rf_forecast, model_name="Random Forest")

# Шаг 5: Визуализация результатов
# -------------------------------

plt.figure(figsize=(12, 6))
plt.plot(y_test, label="Истинные значения", color="blue")
plt.plot(lstm_forecast_flat, label="Прогноз LSTM", color="orange")
plt.plot(rf_forecast, label="Прогноз Random Forest", color="green")
plt.title("Сравнение прогнозов")
plt.xlabel("Наблюдения")
plt.ylabel("Продажи")
plt.legend()
plt.grid()
plt.show()

# Шаг 6: Прогноз на 7 дней 2025 года для каждого бренда
# -----------------------------------------------------

# Получение уникальных брендов
brands = data['Бренд'].unique()

# Словарь для хранения результатов прогнозов по брендам
brand_forecasts = {}

for brand in brands:
    # Фильтрация данных для текущего бренда
    brand_data = data[data['Бренд'] == brand]

    # Подготовка данных
    brand_categorical_features = encoder.transform(brand_data[["Промо", "Праздник", "Сезонность"]]).toarray()
    brand_numeric_features = brand_data[["Цена", "Единицы", "Объем продаж"]].values
    brand_processed_data = np.hstack((brand_numeric_features, brand_categorical_features))
    brand_processed_data = scaler.transform(brand_processed_data)

    # Подготовка последних данных для прогноза
    brand_latest_data = brand_processed_data[-time_steps:]
    brand_forecast_sequence = brand_latest_data[np.newaxis, :, :]

    # Генерация прогноза с использованием LSTM
    brand_future_forecast_lstm = []
    for _ in range(7):
        lstm_pred = lstm_model.predict(brand_forecast_sequence)
        brand_future_forecast_lstm.append(lstm_pred[0, 0])

        # Обновляем входные данные: убираем первый шаг и добавляем новый прогноз
        new_input = np.vstack((
            brand_forecast_sequence[0, 1:],  # Убираем первую строку
            np.hstack((brand_forecast_sequence[0, -1, :-1], lstm_pred[0, 0]))  # Добавляем прогноз
        ))
        brand_forecast_sequence = new_input[np.newaxis, :, :]

    brand_future_forecast_lstm = np.array(brand_future_forecast_lstm)

    # Генерация итогового прогноза с использованием Random Forest
    brand_rf_future_input = np.hstack((
        brand_future_forecast_lstm.reshape(-1, 1),
        encoder.transform(brand_data[["Промо", "Праздник", "Сезонность"]].iloc[:7]).toarray()
    ))
    brand_final_future_forecast = rf_model.predict(brand_rf_future_input)

    # Сохранение прогноза в словарь
    brand_forecasts[brand] = brand_final_future_forecast

    # Визуализация прогноза для бренда
    plt.figure(figsize=(10, 5))
    plt.plot(range(1, 8), brand_final_future_forecast, marker="o", label=f"Прогноз для бренда {brand}")
    plt.title(f"Прогноз на 7 дней 2025 года для бренда {brand}")
    plt.xlabel("Дни")
    plt.ylabel("Продажи")
    plt.legend()
    plt.grid()
    plt.show()

# Сохранение прогнозов для всех брендов
brand_forecasts_df = pd.DataFrame(brand_forecasts)
brand_forecasts_df.to_excel("future_forecast_2025_by_brand.xlsx", index=False)

print("Прогноз на 7 дней 2025 года для каждого бренда завершён. Результаты сохранены в 'future_forecast_2025_by_brand.xlsx'.")
```
